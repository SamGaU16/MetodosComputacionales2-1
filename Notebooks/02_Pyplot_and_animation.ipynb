## pyplot and animation

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as ani
```

```python
from matplotlib import rc
rc("animation", html = "jshtml")
```

## pyplot

```python
x = np.linspace(0, 10, 100)
y = np.sin(np.sqrt(x))
```

```python
plt.plot(x, y)
```

```python
fig, ax = plt.subplots(1)
ax.plot(x, y)
```

```python
z = np.sin(x**2)
fig, ax = plt.subplots(2)
ax[0].plot(x, y)
ax[1].plot(x, z)
```

```python
z = np.sin(x**2)
fig, ax = plt.subplots(2,2)
ax[0, 0].plot(x, y)
ax[0, 1].plot(x, z)
ax[1, 0].plot(x, y)
ax[1, 1].plot(x, z)
ax[1, 1].set_xlabel("x dim")
```

## animation

```python
x = np.linspace(0, 10, 100)
y = np.sin(np.sqrt(x))
fig, ax = plt.figure(fig_size = (6,6))

def update(N):
  for i in range(N):
    ax.plot(x[:i], y[:i])

ani.FuncAnimation(fig, func = update, frames=100)
```
